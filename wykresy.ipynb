```python
def simplify(df):
    base = df[df["Wariant"] == "Wersja bazowa"]
    log = df[df["Wariant"] == "Wersja z logarytmiczną transformacją zmiennej celu"]
    strojenie = df[df["Wariant"] == "Strojenie hiperparametrów"]
    try:
        strojenie_log = df[df["Wariant"] == "Strojenie hiperparametrów po logarytmizacji"]
        best_strojenie = strojenie if strojenie["R2"].values[0] >= strojenie_log["R2"].values[0] else strojenie_log
    except:
        best_strojenie = strojenie  # dla regresji liniowej nie mamy strojenia po logarytmizacji

    best_base = base if base["R2"].values[0] >= log["R2"].values[0] else log
    pca = df[df["Wariant"] == "Wersja po analizie głównych składowych (PCA)"]

    return best_base, best_strojenie, pca
```

```python
models = {
    "MLP": mlp_results,
    "KNN": knn_results,
    "SVR": svr_results,
    "XGBoost": xgboost_results,
    "Gradient Boosting": gradient_boosting_results,
    "Random Forest": random_forest_results,
    "Linear Regression": linear_regression_fixed
}

mae = {"Bazowe": [], "Strojenie": [], "PCA": []}
mape = {"Bazowe": [], "Strojenie": [], "PCA": []}
rmse = {"Bazowe": [], "Strojenie": [], "PCA": []}

for name, df in models.items():
    base, strojenie, pca = simplify(df)

    mae["Bazowe"].append(base["MAE"].values[0])
    mae["Strojenie"].append(strojenie["MAE"].values[0])
    mae["PCA"].append(pca["MAE"].values[0])

    mape["Bazowe"].append(base["MAPE"].values[0])
    mape["Strojenie"].append(strojenie["MAPE"].values[0])
    mape["PCA"].append(pca["MAPE"].values[0])

    rmse["Bazowe"].append(base["RMSE"].values[0])
    rmse["Strojenie"].append(strojenie["RMSE"].values[0])
    rmse["PCA"].append(pca["RMSE"].values[0])
```
